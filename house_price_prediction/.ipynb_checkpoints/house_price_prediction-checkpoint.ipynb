## Работа с пропусками

```python
# посмотрим на картину с пропусками на тепловой карте 
sns.heatmap(df.isnull(), cbar = False)
```

```python
# так как пропусков много, рассмотрим отдельно те столбцы в которых отсутствует хотя бы одно значение
# выведем процент и количество пропущенных значений
for label in df.columns:
    pct_missing = np.mean(df[label].isnull())
    sum_missing = df[label].isnull().sum()
    if sum_missing > 0:
        print('{} - {}% ({})'.format(label, round(pct_missing*100), sum_missing))
```
